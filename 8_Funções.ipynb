# Funções

Podemos definir funções no `Python`. São muito úteis porque permitem:

- reutilizar código
- decompor uma tarefa complexa numa série de tarefas mais simples
- facilitar a leitura e futuras modificações ao programa

```
def nome(arg1, arg2, ...): # cria um objeto do tipo função e atribui-lhe um nome
	'Texto de documentação'
	<instruções>
	return <resultado> # devolve os resultados para a função quando é chamada no programa principal
```

Para chamar uma função, basta escrever o seu nome, colocando os seus argumentos.

## Definir uma Função

Criar a função uma função produto:

````python
def produto(x,y):
  return x*y

#print(produto(2,3))
print(produto('a', 4))
````

## Documentar uma Função

As funções definidas em `Python` podem conter texto de ajuda para o utilizador:

````python
def produto(x,y):
  "Função que recebe dois números e devolve o seu produto"
  return x*y

help(produto) # obter a documentação
````

## Parâmetros das Funções
Todos os exemplos que vimos até agora são de parâmetros posicionais. A ordem pelo qual estes parâmetros são dados à função é importante no caso das operações comutativas, por exemplo:

````python
def operation(x,y,z):
	return x / y + z

print(operation(3, 5, 2))
````

## Parâmetros posicionais

São aqueles que já temos visto, em que colocamos os argumentos pela mesma ordem em que eles estão definidos na função:

````python
print(operation(3, 2, 5))
````

## Parâmetros chave-valor

Podemos mudar a ordem pela qual colocamos os argumentos, se colocarmos as chaves com as quais eles estão definidos na função:

````python
def operation(x,y,z):
	return x / y + z

print(operation(z = 5, x = 3, y = 2))
print(operation(3,2,5))
````

## Mistura de parâmetros

Podemos também fazer uma **mistura dos dois tipos**, mas nesse caso os parâmetros posicionais têm que ser colocados primeiro que os chave-valor:

````python
print(operation(3, z = 5, y = 2))
````

## Valores por omissão

Podemos especificar valores que um parâmetro pode tomar caso o utilizador não o declare quando chama a função:

````python
def soma_iva(valor, iva = 23): # aqui estamos a dizer que, por default, o valor do iva é 23%
	return valor * (1 + iva/100.)

#print(soma_iva(100)) # usa o valor default

print(soma_iva(100, 6)) # usa o valor iva = 6% em vez do valor default de 23%
````

## Alcance dos Nomes

Quando é invocado um objeto numa função, como parâmetro/argumento, ou no seu interior, a resolução dos nomes segue as seguintes regras:



### 1. Procura no espaço dos nomes local da função

Procura nos nomes que foram atribuídos no interior da função ou na sua definição (através dos parâmetros de entrada).

Ex: variável dentro da função.

````python
from math import pi

def assign():
	pi = 3.141592
	return pi

pi = 3

#print(assign()) # o pi da função é independente do resto do programa
print(pi) # o programa principal não consegue aceder ao pi do interior da função
````

### 2. Procura no espaço global

Se o nome não for encontrado no espaço local, procura no módulo principal.

Ex: variável global do resto do programa.

````python
from math import pi

def assign():
	return pi

pi = 3

#print(assign()) # como o pi não está definido dentro da função, ele procura-o no programa global
print(pi)
````

### 3. Procura no espaço de nomes pré-definidos

Caso não exista, faz uma procura nos nomes pré-definidos do `Python`.

````python
from math import pi

def assign():
	return pi

print(assign()) # como não há pi definido em lado nenhum, ele vai aos nomes pré-definidos no python
````

### 4. Erro

Caso não seja encontrado, será levantado um erro do tipo `NameError: name is not defined`.

````python
del pi

def assign():
	return pi

print(assign()) # como não há pi definido em lado nenhum, ele vai aos nomes pré-definidos no python
````

## 2. Contar caracteres numa string

_(Ficha 4, Problema 1)_

Utilize essa função num programa que dada uma `string` introduzida pelo utilizador dê como resultado o número de vezes que cada caracter aparece na `string`. O resultado deve aparecer na forma de um dicionário que tem como chaves cada carater da palavra e como valores associados, o número de vezes que esse carater aparece na palavra.

````python
#@title Solução

def conta_c(string, caracter):
    contador = 0                # inicializar o contador
    for letra in string:        # Para cada letra na string:
        if letra == caracter:   # Se a letra for igual ao caracter
            contador += 1      # Adicionar uma unidade ao contador
    return contador             # Devolve o resultado do contador

frase = input('Introduza a frase: ')

contagem = {} # dicionário vazio

for char in frase:      # Para cada caracter na frase introduzida:
    contagem[char] = conta_c(frase, char) # adicionar o novo elemento ao resultado

print(contagem)
````

## 3. Raiz Digital

_(Ficha 4, Problema 2)_

A raiz digital de um número inteiro não negativo é a soma dos dígitos que formam o número. Caso esta soma tenha mais do que um dígito, repete-se o processo, uma ou mais vezes, até que o resultado seja um só dígito.

- **Exemplo:** a raiz digital de `9299` é `2`:
  1. 9 + 2 + 9 + 9 = `29`
  2. 2 + 9 = `11`
  3. 1 + 1 = `2`

Escreva uma função `digital` que calcule a raiz digital de um número inteiro não negativo e utilize-a num programa que peça ao utilizador um conjunto de `N` números e dê a raiz digital desses números.

````python
#@title Solução

def digital(n):
    if int(n) > 0:              # Se for um número inteiro não negativo:
        while len(n) > 1:           # Enquanto n não tem um dígito:
            soma = 0                # inicia a soma como 0
            for algarismo in n:     # Para cada algarismo no numero n:
                soma = soma + int(algarismo)    # soma o próximo algarismo
            n = str(soma)           # no fim, atualiza n como a soma final
            print(n)                # imprime o n intermédio de cada passo do while
        return n                    # imprime o n final que é a raiz digital
    else:
        return 'não definida.'

N = int(input('Quantos números quer introduzir? '))

for i in range(N):
    n = input('Introduza o número: ')
    print('A raiz digital de', n, 'é', digital(n))
````

## 4. Soma de uma lista de números

_(Ficha 4, Problema 4)_

Escreva um programa que defina uma função `soma(L)` que calcula a soma dos elementos de uma lista de números. Caso a lista seja vazia a função deve devolver o valor `0`.
Utilize a função no programa, usando como argumento uma lista com n números aleatórios entre `-100` e `100`, sendo `n` um inteiro definido pelo utilizador.

````python
#@title Solução

def soma(L):
    if len(L) == 0: # Se a lista estiver vazia devolver 0
        return 0
    else:           # Senão, calcula a soma:
        soma = 0
        for i in L:
            soma += i
        return soma

from random import randint

n = int(input('Quantos números quer gerar? '))

aleatorios = []     # Lista vazia para incluir os numeros aleatorios

for i in range(n):  # Para cada elemento entre 0 e n:
    aleatorios.append(randint(-100, 100))   # gerar um novo aleatório e adicioná-lo à lista

print('A lista gerada é', aleatorios)
print('A soma da lista é', soma(aleatorios))
````

Defina uma função `Riemann(n,p)` que lhe permita obter os n primeiros elementos da série de Riemann:

$$S(n,p) = \sum_{i=0}^n \frac{1}{i^p}$$

A função deve devolver o valor da série de Riemann para os valores `n` e `p` introduzidos na chamada da função. O valor de `n` deve ser inteiro e o de `p` pode ser real.

O número de elementos dessa lista deve ser igual ao número de sequências que existem para `n`. Cada elemento dessa lista deve ser uma tupla onde o 1º elemento é o 1º número da sequência e o 2º elemento, o último da mesma sequência. Por exemplo, `sequencias(9)` deve devolver a lista `[(2, 4), (4, 5)]`.

````python
#@title Solução

def sequencias(n):
    seq = []                        # lista vazia que vamos preencher com os resultados
    for i in range(1, n):           # numero por onde vamos comecar a somar
        soma = 0
        for j in range(i, n):       # para cada numero j entre i e n:
            soma = soma + j         # somar o proximo numero da sequencia
            if soma == n:           # se a soma ficar igual ao valor que pretendemos:
                seq.append((i, j))  # adicionar os numeros que verificam a soma
    return seq                      # devolver a lista com todas as sequências


n = int(input('Qual é o número inteiro que pretende avaliar? '))

print('As sequências para este número são:', sequencias(n))
````

É, no entanto, possível encontrar vários conjuntos de quatro números inteiros maiores que a unidade que satisfazem a relação:

$$a^3+b^3+c^3=d^3$$

Por exemplo:
- $3^3+4^3+5^3=6^3$
- $7^3+14^3+17^3=20^3$

### a) Função `cubos()`

Escreva uma função `cubos` que encontre, caso existam, o conjunto de números $(a, b, c)$ que satisfazem a igualdade acima para um dado valor $d$ (parâmetro de entrada da função). A função deve devolver uma tupla com os valores $a$, $b$ e $c$ ou uma tupla vazia, caso estes valores não existam.

````python
#@title Solução

def cubos(d):
    """ Função que calcula os cubos de um número"""
    resultado = []      # lista com os resultados vazia
    for a in range(1,d):    # para cada a entre 1 e d:
        for b in range(a,d):    # para cada b entre a e d:
            for c in range(b,d):    # para cada c entre b e d:
                parcela = a**3 + b**3 + c**3
                if parcela == d**3:  # verificar a condição
                    # se a condição se verificar, adicionar a,b,c à lista
                    resultado.append(a)
                    resultado.append(b)
                    resultado.append(c)
    return tuple(resultado)

d = int(input('Introduza d: '))
print('O resultado é', cubos(d))
````

## 9. Números amigos

_(Ficha 5, Problema 7)_

Dois números dizem-se "amigos" se a soma dos divisores de um deles for igual ao outro e vice-versa.

- **Exemplo** para `220` e `284`
  - Divisores de `220`: `1`, `2`, `4`, `5`, `10`, `11`, `20`, `22`, `44`, `55` e `110`, cuja soma é igual a `284`.
  - Divisores de `284`: `1`, `2`, `4`, `71` e `142`, cuja soma é igual a `220`.

### a) Função `divisores()`

Escreva uma função que dê como resultado uma lista que contenha os divisores de um número inteiro, `n`.

- **Exemplo**:
  ```
  print (divisores(9))
  ```
  Deve devolver `[1, 3]`

````python
#@title Solução

def divisores(n):
    div = []                # lista com os divisores
    for i in range(1,n):    # para cada número até n
        if n % i == 0:      # verificar se é divisor de n e adicionar à lista se sim
            div.append(i)
    return div

n = int(input('Introduza um número n: '))
print(divisores(n))
````

### b) Procura de amigos

Utilize esta função num programa que procura os números amigos até `N` (sendo `N` um número inteiro introduzido pelo utilizador).
- **Exemplo:**
  ```
  Introduza N: 2000
  ```
  Deve devolver `[(220, 284), (1184, 1210)]`

````python
#@title Solução

N = int(input('Introduza um número inteiro: '))

amigos = []

for conhecido_1 in range(1, N + 1): # para cada numero de 1 a N:

    divisores_1 = divisores(conhecido_1)     # Determinar os seus divisores
    conhecido_2 = sum(divisores_1)           # somar os seus divisores

    divisores_2 = divisores(conhecido_2)     # determinar os divisores da sua soma

    if sum(divisores_2) == conhecido_1 and conhecido_1 < conhecido_2:  # se a soma dos divisores do segundo numero for igual ao primeiro: (a segunda condição é para evitar repetição dos numeros amigos)
        amigos.append((conhecido_1, conhecido_2))                       # temos um número amigo

print(amigos)
````

## 10. Cálculos de combinatória

_(Ficha 5, Problema 8)_


### a) Função `fatorial()`

Escreva uma função `fatorial(n)` para calcular o fatorial de um número inteiro e não negativo `n`.

````python
#@title Solução

def fatorial(n):
    """ Devolver o fatorial do número n
    """
    if n == 0:
        return 1
    elif (n < 0) or (type(n) != int): # se for um número negativo ou não inteiro, não devolve nada
        return None
    else:
        fatorial = 1
        for i in range(1, n+1):
            fatorial = fatorial * i
        return fatorial

n = int(input('Introduza um valor para n: '))
print('O fatorial de n é ', fatorial(n))
````

### b) Função `combinations()`

Utilize a função anterior na função `combinations(n,p)` para calcular as combinações:

$$\binom{n}{p} = \frac{n!}{(n-p)! \cdot p!}$$

Se $n < p$, a função `combinations(n,p)` deve devolver o valor `-1`.

````python
#@title Solução

def combinations(n, p):
    if n < p:           # se n < p, devolver -1
        return -1
    else:               # caso contrário, aplicar a fórmula
        return fatorial(n) / (fatorial(n-p) * fatorial(p))


n = int(input('Introduza n: '))
p = int(input('Introduza p: '))

print(f'comb({n}, {p}) = ', combinations(n, p))
````

- **Nota:** Neste caso, a primeira `string` tem 5 palavras, a segunda tem 6 palavras, e o programa compara as palavras da `string` `a` com as primeiras 5 palavras da `string` `b`.

````python
#@title Solução

def separa(frase):
    palavras = []
    inicio = 0      # inicio da primeira palavra

    for posicao in range(len(frase)):   # para cada posição na frase:
        if frase[posicao] == ' ':       # se encontrarmos um espaço, estamos numa palavra nova
            palavras.append(frase[inicio:posicao])    # colocamos a última palavra na lista (começa no inicio e acaba na posição em que estamos)
            inicio = posicao + 1           # atualizar o início da próxima palavra

    palavras.append(frase[inicio:])     # adicionar a última palavra, que não tem um " " no final
    return palavras     # devolver a lista de palavras

def compara(lista_1, lista_2):

    res = []    # lista com o resultado

    for (palavra_1, palavra_2) in zip(lista_1, lista_2):        # para cada palavra das listas
        letras_iguais = 0                                       # incializa o nº de letras iguais
        for (letra_1, letra_2) in zip(palavra_1, palavra_2):    # para cada letra nas palavras
            if letra_1 == letra_2:                              # se as letras forem iguais
                letras_iguais += 1                              # acrescenta 1 ao contador
        res.append(letras_iguais)                               # no fim da palavra, adicionamos o resultado
    return res

frase_1 = input('Introduza a primeira frase: ')
frase_2 = input('Introduza a segunda frase: ')

print(compara(separa(frase_1), separa(frase_2)))                # chamar a função com as strings pretendidas
````
